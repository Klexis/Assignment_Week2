```python
%matplotlib inline
houses[1].plot(kind='hist')
```

The histogram is so ugly because it is skewed by huge intensity. We can transform the x-axis into corresponding logarithmic values in order to make the plot more informative.

```python
import numpy as np
import matplotlib.pyplot as plt
dfs = np.log(pd.Series.tolist(houses[1]))/np.log([10])
plt.hist(dfs, bins=100)
```

```python
#dfs = np.log(pd.Series.tolist(houses["LONDON" in houses[13]][1]))/np.log([10])
#houses.loc[houses[12]=='OXFORD',:]
Bool = []
for names in houses[13]:
    if 'LONDON' in names:
        Bool.append(True)
    else:
        Bool.append(False)
```

```python
Bool = pd.Series(Bool)
```

```python
houses_1 = houses[Bool]
```

```python
Bool = []
for names in houses[13]:
    if 'LONDON' in names:
        Bool.append(False)
    else:
        Bool.append(True)
houses_2 = houses[Bool]
```

```python
dfs = np.log(pd.Series.tolist(houses_1[1]))/np.log([10])
plt.hist(dfs, bins=100)
```

```python
dfs = np.log(pd.Series.tolist(houses_2[1]))/np.log([10])
plt.hist(dfs, bins=100)
```

Yes, real estate is more expensive in London since the histogram of the rates of real estate in London has been shifted to the right.
